```python
# Import
import pandas as pd
```

### Load CSV

```python
# Load CSV
path = '../data/gsearch_jobs.csv'
data = pd.read_csv(path)

# Preview dataset
data.head()
```

```python
# View columns to determine which columns are necessary to keep
data.columns
```

```python
# Only keep necessary columns
data_clean = data[['title', 'company_name', 'location', 'via', 'description', 'extensions', 'schedule_type', 'work_from_home', 'salary_hourly', 'salary_yearly', 'salary_standardized', 'description_tokens']]
data_clean.head()
```

```python
data_clean.info()
```

```python
from json import loads, dumps
import json
```

```python
# Export clean data to json file

result = data_clean.to_json(orient="records")
parsed = loads(result)
```

```python
with open('../data/clean_data.json', 'w') as fp:
    json.dump(parsed, fp)   
```
